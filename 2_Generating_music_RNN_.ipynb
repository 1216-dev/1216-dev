```python
import os
import json
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import LSTM, Dropout, Dense, Activation, Embedding
```

```python
data_directory = r"C:\Users\sweta\OneDrive\Desktop\DL Projects\Music generation\char-rnn-keras-master\data/"
data_file = r"input.txt"
charIndex_json = "char_to_idx.json"
model_weights_directory = r'C:\Users\sweta\OneDrive\Desktop\DL Projects\Music generation\model/'
BATCH_SIZE = 16
SEQ_LENGTH = 64
```

```python
def make_model(unique_chars):
    model = Sequential()

    model.add(Embedding(input_dim = unique_chars, output_dim = 512, batch_input_shape = (1, 1)))

    model.add(LSTM(256, return_sequences = True, stateful = True))
    model.add(Dropout(0.2))

    model.add(LSTM(256, return_sequences = True, stateful = True))
    model.add(Dropout(0.2))

    model.add(LSTM(256, stateful = True))
    #remember, that here we haven't given return_sequences = True because here we will give only one character to generate the
    #sequence. In the end, we just have to get one output which is equivalent to getting output at the last time-stamp. So, here
    #in last layer there is no need of giving return sequences = True.
    model.add(Dropout(0.2))

    model.add((Dense(unique_chars)))
    model.add(Activation("softmax"))

    return model
```

```python
def generate_sequence(epoch_num, initial_index, seq_length):
    with open(os.path.join(data_directory, charIndex_json)) as f:
        char_to_index = json.load(f)
    index_to_char = {i:ch for ch, i in char_to_index.items()}
    unique_chars = len(index_to_char)

    model = make_model(unique_chars)
    model.load_weights(model_weights_directory + "weights.{}.h5".format(epoch_num))

    sequence_index = [initial_index]

    for _ in range(seq_length):
        batch = np.zeros((1, 1))
        batch[0, 0] = sequence_index[-1]

        predicted_probs = model.predict_on_batch(batch).ravel()
        sample = np.random.choice(range(unique_chars), size = 1, p = predicted_probs)

        sequence_index.append(sample[0])

    seq = ''.join(index_to_char[c] for c in sequence_index)

    cnt = 0
    for i in seq:
        cnt += 1
        if i == "\n":
            break
    seq1 = seq[cnt:]
    #above code is for ignoring the starting string of a generated sequence. This is because we are passing any arbitrary
    #character to the model for generating music. Now, the model start generating sequence from that character itself which we
    #have passed, so first few characters before "\n" contains meaningless word. Model start generating the music rhythm from
    #next line onwards. The correct sequence it start generating from next line onwards which we are considering.

    cnt = 0
    for i in seq1:
        cnt += 1
        if i == "\n" and seq1[cnt] == "\n":
            break
    seq2 = seq1[:cnt]
    #Now our data contains three newline characters after every tune. So, the model has leart that too. So, above code is used for
    #ignoring all the characters that model has generated after three new line characters. So, here we are considering only one
    #tune of music at a time and finally we are returning it..

    return seq2
```

```python
ep = int(input("1. Enter the epoch number of the model you want to load. Small number will generate more errors in music: "))
ar = int(input("\n2. Enter any number between 0 to 86 which will be given as initial charcter to model for generating sequence: "))
ln = int(input("\n3. Enter the length of music sequence you want to generate. Typical number is between 300-600. Too small number will generate hardly generate any sequence: "))

music = generate_sequence(ep, ar, ln)

print("\nMUSIC SEQUENCE GENERATED: \n")

print(music)
```

```python
import ipywidgets as widgets
from IPython.display import display, Audio, clear_output
import os

class MusicPlayerWidget(widgets.VBox):
    def __init__(self):
        super().__init__()

        self.playlist = []
        self.current_track_index = None
        self.audio_widget = None

        # Header
        header = widgets.HTML("<h2 style='color: #4CAF50;'>🎵 Music Player 🎵</h2>")

        # Play, Pause, Stop buttons with styling
        self.play_button = widgets.Button(
            description="Play",
            button_style='success',
            layout=widgets.Layout(width='80px', height='40px')
        )
        self.pause_button = widgets.Button(
            description="Pause",
            button_style='warning',
            layout=widgets.Layout(width='80px', height='40px')
        )
        self.stop_button = widgets.Button(
            description="Stop",
            button_style='danger',
            layout=widgets.Layout(width='80px', height='40px')
        )

        # Add and Delete buttons
        self.add_button = widgets.Button(
            description="Add",
            button_style='info',
            layout=widgets.Layout(width='80px', height='40px')
        )
        self.del_button = widgets.Button(
            description="Delete",
            button_style='info',
            layout=widgets.Layout(width='80px', height='40px')
        )

        # Playlist management
        self.playlist_widget = widgets.Select(
            options=self.playlist,
            description='Playlist',
            layout=widgets.Layout(width='300px', height='150px')
        )

        # Output area
        self.output = widgets.Output()
        self.status_label = widgets.Label(value="Welcome to the Python Music Player")

        # Bind button clicks to actions
        self.play_button.on_click(self.on_play_button_clicked)
        self.pause_button.on_click(self.on_pause_button_clicked)
        self.stop_button.on_click(self.on_stop_button_clicked)
        self.add_button.on_click(self.on_add_button_clicked)
        self.del_button.on_click(self.on_del_button_clicked)

        # Assemble UI elements
        self.children = [
            header,
            self.playlist_widget,
            self.play_button,
            self.pause_button,
            self.stop_button,
            self.add_button,
            self.del_button,
            self.status_label,
            self.output
        ]

    def on_play_button_clicked(self, b):
        if self.playlist_widget.value:
            track_path = self.playlist_widget.value
            with self.output:
                clear_output(wait=True)
                self.audio_widget = Audio(track_path, autoplay=True)
                display(self.audio_widget)
                self.status_label.value = f"Playing {os.path.basename(track_path)}"

    def on_pause_button_clicked(self, b):
        self.status_label.value = "Pause functionality is not available in this widget"

    def on_stop_button_clicked(self, b):
        if self.audio_widget:
            with self.output:
                clear_output(wait=True)
                self.audio_widget = None
                self.status_label.value = "Stopped playback"

    def on_add_button_clicked(self, b):
        # For simplicity, this example uses a fixed path; in practice, use a file dialog
        # You can replace this with a file dialog or other input method in a notebook environment
        file_path = '/content/output using weight.100.h5.mp3'
        if file_path and file_path not in self.playlist:
            self.playlist.append(file_path)
            self.playlist_widget.options = self.playlist
            self.status_label.value = f"Added {os.path.basename(file_path)}"

    def on_del_button_clicked(self, b):
        if self.playlist_widget.value:
            self.playlist.remove(self.playlist_widget.value)
            self.playlist_widget.options = self.playlist
            self.status_label.value = f"Removed {os.path.basename(self.playlist_widget.value)}"

# Usage example
music_player = MusicPlayerWidget()

# Display the music player widget
display(music_player)
```
